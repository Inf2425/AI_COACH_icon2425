## SOTTOFASE 2 – Uniformazione dei nomi dei giocatori nel dataset.

Obiettivo:

Rendere uniforme la colonna Player in tutti i file:

* Trasformazione in minuscolo
* Rimozione di spazi superflui
* Rimozione accenti e caratteri speciali
* Rinomina uniforme della colonna se ha nomi diversi (es. Name, Giocatore, ecc.)

 Step 1: Funzione di normalizzazione del nome

```python
import unicodedata

def normalizza_nome(nome):
    if pd.isnull(nome):
        return ''
    nome = str(nome).lower().strip()
    nome = unicodedata.normalize('NFKD', nome).encode('ascii', errors='ignore').decode('utf-8')
    return nome
```

Step 2: Applica la normalizzazione su tutti i file

* Rinominare correttamente la colonna (se necessario).
* Applicare la funzione normalizza_nome().

```python
for file_name, df in dfs.items():
    # Trova e rinomina la colonna 'Player' se ha nomi alternativi
    possible_names = ['Player', 'player', 'Giocatore', 'Name']
    colonna_player = next((col for col in df.columns if col in possible_names), None)

    if colonna_player and colonna_player != 'Player':
        df.rename(columns={colonna_player: 'Player'}, inplace=True)

    if 'Player' in df.columns:
        df['Player'] = df['Player'].apply(normalizza_nome)
        print(f" Colonna 'Player' normalizzata in {file_name}")
    else:
        print(f" Colonna 'Player' NON trovata in {file_name}")
```

## SOTTOFASE 4 - Merge progressivo su Player.

 Obiettivo

Unificare tutti i DataFrame presenti in dfs in un unico dataset (df_merged) che:

* Ha una riga per ogni giocatore
* Include tutte le metriche pulite delle sottofasi precedenti
* Usa la colonna 'Player' come chiave per il join

Strategia di join

Usiamo un merge esterno (outer) così nessun giocatore viene escluso se non è presente in tutti i CSV.

Uniamo progressivamente tutti i DataFrame.



```python
from functools import reduce
pd.set_option('future.no_silent_downcasting', True)

# Step 1: Rinomina i campi 'Country' per evitare conflitti
dataframes = []
country_cols = []

for filename, df in dfs.items():
    if df is not None and 'Player' in df.columns:
        df = df.copy()
        if 'Country' in df.columns:
            suffix = filename.replace('.csv', '').replace('player_', '').replace('elenco_giocatori_per_squadre_2023_2024', 'squadre')
            new_country_col = f'Country_{suffix}'
            df = df.rename(columns={'Country': new_country_col})
            country_cols.append(new_country_col)
        dataframes.append(df)

# Step 2: Merge progressivo su 'Player'
df_merged = reduce(lambda left, right: pd.merge(left, right, on='Player', how='outer'), dataframes)

# Step 3: Crea una nuova colonna 'Country' con il primo valore non nullo tra tutti i Country_*
temp = df_merged[country_cols].bfill(axis=1)
temp = temp.infer_objects(copy=False)
df_merged['Country'] = temp.iloc[:, 0]


# Step 4: Rimuovi tutte le colonne 'Country_*'
df_merged = df_merged.drop(columns=country_cols)

# Step 5: Output
print(f" Merge completato: {df_merged.shape[0]} giocatori, {df_merged.shape[1]} colonne")
print(df_merged.head(10))  # puoi usare anche display(df_merged) se sei in Jupyter/Colab
```

## SOTTOFASE 5 – Gestione NaN

```python
# Sostituiamo i NaN in tutte le colonne tranne 'Country'
colonne_da_riempire = df_merged.columns.difference(['Country'])
df_merged[colonne_da_riempire] = df_merged[colonne_da_riempire].fillna(0)

# Controllo: stampiamo eventuali colonne che ancora contengono NaN (eccetto Country)
nan_residui = df_merged.drop(columns='Country').isna().sum()
nan_residui = nan_residui[nan_residui > 0]

if nan_residui.empty:
    print("Tutti i NaN (tranne 'Country') sono stati sostituiti con 0.")
else:
    print("Ancora presenti NaN (escluso 'Country') nelle seguenti colonne:")
    print(nan_residui)
```
